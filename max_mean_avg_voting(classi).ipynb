```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
x,y = make_moons(n_samples=1000,noise=0.2)
x,y
```

```python
df = {"data1":x[:,0],"data2":x[:,1],"output":y}
```

```python
dataset = pd.DataFrame(df)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset,hue=y)
plt.show()
```

```python
x_a = dataset.iloc[:,:-1]
y_a =  dataset["output"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_a,y_a,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt.score(x_train,y_train)*100,dt.score(x_test,y_test)*100
```

```python
sv = SVC()
sv.fit(x_train,y_train)
sv.score(x_train,y_train)*100,sv.score(x_test,y_test)*100
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
gnb.score(x_train,y_train)*100,gnb.score(x_test,y_test)*100
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
li = [("dt1",DecisionTreeClassifier()),("sv1",SVC()),("gnb1",GaussianNB())]
```

```python
vc = VotingClassifier(li)
vc.fit(x_train,y_train)
```

```python
vc.score(x_train,y_train)*100,vc.score(x_test,y_test)*100
```
